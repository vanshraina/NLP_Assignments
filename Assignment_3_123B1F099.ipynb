```python
# Import required modules
import re
from collections import Counter, defaultdict

# Sample text
data_text = """Machine learning and natural language processing are important fields in artificial intelligence.
Natural language techniques help computers understand human communication effectively."""

# Choose value of N (Bigram = 2, Trigram = 3)
n_value = 3

# Input sequence
user_input = "natural language"
```

```python
# Step 1: Clean the text
clean_data = data_text.lower()
clean_data = re.sub(r'[^a-z\s]', '', clean_data)

word_tokens = clean_data.split()


# Step 2: Build N-gram model
model = defaultdict(Counter)

for index in range(len(word_tokens) - n_value + 1):
    prev_words = tuple(word_tokens[index:index + n_value - 1])
    next_w = word_tokens[index + n_value - 1]
    model[prev_words][next_w] += 1


# Step 3: Convert counts to probabilities
probability_model = {}

for context_key, next_words in model.items():
    total = sum(next_words.values())
    probability_model[context_key] = {}

    for word, count in next_words.items():
        probability_model[context_key][word] = count / total
```

```python
# Step 4: Predict next word
user_input = user_input.lower()
user_input = re.sub(r'[^a-z\s]', '', user_input)

input_list = user_input.split()
context_input = tuple(input_list[-(n_value - 1):])

print("\nContext Given:", " ".join(context_input))

if context_input in probability_model:
    next_word = max(probability_model[context_input],
                    key=probability_model[context_input].get)
    print("Next Word Prediction:", next_word)
else:
    print("Prediction not possible for given input")
```
