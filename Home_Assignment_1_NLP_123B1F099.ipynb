```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer


# STEP 0: INITIALIZATION
print("\n" + "="*70)
print("STEP 0: INITIALIZING NLTK RESOURCES")
print("="*70)

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')


# PREPROCESS FUNCTION
def advanced_preprocess(text):

    print("\n" + "="*70)
    print("TEXT PREPROCESSING STEPS")
    print("="*70)

    # Step 1: Lowercase
    lower_text = text.lower()
    print(f"[1] Lowercase:\n{lower_text}")

    # Step 2: Noise Removal
    clean_text = re.sub(r'[^a-zA-Z\s]', '', lower_text)
    print(f"\n[2] Noise Removed:\n{clean_text}")

    # Step 3: Tokenization
    tokens = word_tokenize(clean_text)
    print(f"\n[3] Tokens:\n{tokens}")

    # Step 4: Stopword Removal
    stop_words = set(stopwords.words('english'))
    filtered = [w for w in tokens if w not in stop_words]
    print(f"\n[4] Stopwords Removed:\n{filtered}")

    # Step 5: Lemmatization
    lemmatizer = WordNetLemmatizer()
    lemmas = [lemmatizer.lemmatize(w) for w in filtered]
    print(f"\n[5] Lemmatized Output:\n{lemmas}")

    return lemmas


# STEP 1: EXECUTION
print("\n" + "="*70)
print("STEP 1: RUNNING PREPROCESSING")
print("="*70)


sample = "Artificial Intelligence systems are rapidly evolving, and researchers are exploring innovative NLP techniques every day!"

final_output = advanced_preprocess(sample)


# FINAL OUTPUT
print("\n" + "="*70)
print("FINAL OUTPUT")
print("="*70)
print(final_output)
```
