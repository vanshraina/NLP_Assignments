```python
# Import required libraries
import nltk
import re
from nltk.tokenize import sent_tokenize as sent_tok, word_tokenize as word_tok
from nltk import pos_tag as pos

# Download required resources
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
# Step 1: Input Paragraph (modified)
paragraph = """The intelligent boy solved a difficult problem.
A beautiful garden has colorful flowers and green trees."""

# List to store noun phrases
noun_phrases = []


# Step 2: Sentence Splitting
sentence_list = sent_tok(paragraph)


# Step 3: Process Each Sentence
for sent in sentence_list:

    # Clean text
    sent = sent.lower()
    sent = re.sub(r'[^a-z\s]', '', sent)

    # Tokenize words
    tokens = word_tok(sent)

    # POS tagging
    tagged = pos(tokens)


    # Step 4: Define Proper NP Grammar
    # Pattern: Determiner + Adjectives + Noun
    np_rule = "NP: {<DT>?<JJ>*<NN|NNS>+}"

    parser = nltk.RegexpParser(np_rule)
    parsed_tree = parser.parse(tagged)


    # Step 5: Extract Noun Phrases
    for subtree in parsed_tree.subtrees():
        if subtree.label() == "NP":
            phrase = " ".join(word for word, tag in subtree.leaves())
            noun_phrases.append(phrase)


# Output
print("\nExtracted Noun Phrases:\n")

for phrase in noun_phrases:
    print(phrase)
```
