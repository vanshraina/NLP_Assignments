```python
# Import necessary modules
import nltk
from nltk.tokenize import word_tokenize as tokenize_words
from nltk.stem import PorterStemmer as Stemmer
from nltk.stem import WordNetLemmatizer as Lemmatizer

# Download required resources (run once)
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet')

# Sample text corpus
sample_text = "The children are playing in the gardens and enjoying their activities."

# Convert text to lowercase
processed_text = sample_text.lower()
```

```python
# Step 1: Tokenization
word_list = tokenize_words(processed_text)


# Step 2: Stemming
stem_tool = Stemmer()
stems = [stem_tool.stem(w) for w in word_list]


# Step 3: Lemmatization
lemma_tool = Lemmatizer()
lemmas = []

for w in word_list:
    lemmas.append(lemma_tool.lemmatize(w))

# Output Results
print("\nTokenized Words:")
print(word_list)

print("\nAfter Stemming:")
print(stems)

print("\nAfter Lemmatization:")
print(lemmas)
```
