```python
# STEP 1: Import Required Libraries
import nltk
from nltk.tokenize import word_tokenize as wt
from nltk import pos_tag as tagger
import re

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
# Sample dataset
paragraph = """Artificial intelligence is transforming industries and improving decision making.
It enables systems to learn from data and perform tasks efficiently.
Applications include speech recognition, recommendation systems, and automation."""


# Step 1: Text Cleaning
clean_text = paragraph.lower()
clean_text = re.sub(r'[^a-z\s]', '', clean_text)


# Step 2: Tokenization
words = wt(clean_text)


# Step 3: POS Tagging
tagged_output = tagger(words)


# Display Output
print("\nPOS Tagging Result:\n")

for w, t in tagged_output:
    print(f"{w} : {t}")
```
