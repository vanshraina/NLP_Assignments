```python
!pip install gensim
import gensim.downloader as api
import numpy as np


# STEP 1: LOAD EMBEDDING MODELS
print("\n" + "="*70)
print("STEP 1: LOADING WORD EMBEDDING MODELS")
print("="*70)

# GloVe
glove = api.load("glove-twitter-25")

# Word2Vec (using larger pretrained vectors)
word2vec = api.load("glove-wiki-gigaword-100")

# FastText
fasttext = api.load("fasttext-wiki-news-subwords-300")

print("All models loaded successfully!")


# FUNCTION TO COMPARE
def compare_all_embeddings(w1, w2):
    print(f"\nComparing: '{w1}' vs '{w2}'")
    print("-" * 50)
    print(f"Word2Vec Similarity : {word2vec.similarity(w1, w2):.4f}")
    print(f"GloVe Similarity    : {glove.similarity(w1, w2):.4f}")
    print(f"FastText Similarity : {fasttext.similarity(w1, w2):.4f}")


# STEP 2: SEMANTIC SIMILARITY TEST
print("\n" + "="*70)
print("STEP 2: SEMANTIC SIMILARITY ANALYSIS")
print("="*70)


compare_all_embeddings("doctor", "nurse")
compare_all_embeddings("teacher", "student")
compare_all_embeddings("car", "truck")


# STEP 3: OOV TEST
print("\n" + "="*70)
print("STEP 3: OUT-OF-VOCABULARY (OOV) TEST")
print("="*70)

test_word = "datasciencepro"

print(f"Testing OOV word: '{test_word}'\n")

# FastText
try:
    ft_vec = fasttext[test_word]
    print(" FastText: SUCCESS (Generated using subword information)")
except:
    print(" FastText: FAILED")

# Word2Vec
try:
    w2v_vec = word2vec[test_word]
    print(" Word2Vec: SUCCESS")
except KeyError:
    print(" Word2Vec: FAILED (Word not in vocabulary)")

# GloVe
try:
    g_vec = glove[test_word]
    print(" GloVe: SUCCESS")
except KeyError:
    print(" GloVe: FAILED (Word not in vocabulary)")

print("\n" + "="*70)
print("ANALYSIS COMPLETE")
print("="*70)
```
